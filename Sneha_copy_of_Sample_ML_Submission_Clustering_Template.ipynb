## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Fill missing values in rating_count with mode
df['rating_count'].fillna(df['rating_count'].mode()[0], inplace=True)

# Remove symbols from price and discount columns
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False)

df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False)

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False)

# Remove commas from rating_count
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '', regex=False)

# Convert columns to numeric datatype
df['discounted_price'] = pd.to_numeric(df['discounted_price'])
df['actual_price'] = pd.to_numeric(df['actual_price'])
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'])
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Verify changes
df.info()
````

### What all manipulations have you done and insights you found?

#### **Manipulations Performed**

* Identified missing values and filled the missing values in the rating_count column using the mode.
* Removed currency symbols (₹), commas (,), and percentage symbols (%) from numerical columns.
* Converted discounted_price, actual_price, discount_percentage, rating, and rating_count into numeric data types.
* Verified the updated data types to ensure the dataset was ready for further analysis.

#### **Insights**

* The dataset contains 1,465 Amazon products with pricing, discounts, ratings, and review information.
* Only the rating_count column had 2 missing values, indicating that the dataset is nearly complete.
* Several numerical columns were stored as text due to currency symbols and commas, so preprocessing was necessary.
* After cleaning and data type conversion, the dataset became suitable for exploratory data analysis and unsupervised learning.

#### Chart - 1

````python
# Chart - 1 visualization code

plt.figure(figsize=(10,6))
sns.histplot(df['discounted_price'], bins=30, kde=True, color='royalblue')

plt.title('Distribution of Discounted Price', fontsize=15)
plt.xlabel('Discounted Price (₹)', fontsize=12)
plt.ylabel('Number of Products', fontsize=12)

plt.show()
````

##### 1. Why did you pick the specific chart?

* A **histogram** is an appropriate choice for visualizing the distribution of a continuous numerical variable like **discounted price**.
* It helps identify the frequency of products across different price ranges, detect skewness in the data, and reveal the presence of outliers.

##### 2. What is/are the insight(s) found from the chart?

* **Most products** are concentrated in the **lower discounted price range**, indicating that **Amazon offers a large number of affordable products**.

* The distribution is **highly right-skewed**, with only a few products having **very high discounted prices**.

* A small number of premium products act as outliers, extending the distribution toward higher prices.

* The wide variation in discounted prices suggests that Amazon caters to customers with different budget preferences.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

**Yes, the insights can create a positive business impact.**

* Since most products fall into the affordable price range, businesses can focus on maintaining sufficient inventory and promotional offers for these products to maximize sales.

* Premium-priced products can be targeted through personalized marketing campaigns aimed at customers seeking high-end items.

* The presence of a few extremely expensive products indicates a diverse product portfolio, helping Amazon serve multiple customer segments.

**Negative Growth Insight:**

* The highly skewed distribution suggests that very expensive products are purchased less frequently. If excessive inventory is maintained for these products, it may lead to higher storage costs and slower inventory turnover. Therefore, inventory planning should be aligned with product demand.
