### Variables Description

* **product_id** :	Unique identifier assigned to each product.

* **product_name** :	Name of the product listed on Amazon.

* **category** :	Category to which the product belongs (e.g., Electronics, Home & Kitchen).

* **discounted_price** :	Selling price of the product after applying the discount.

* **actual_price** :	Original price of the product before the discount.

* **discount_percentage** : 	Percentage of discount offered on the product.

* **rating** :	Average customer rating given to the product (typically out of 5).

* **rating_count** :	Total number of customers who have rated the product.

* **about_product** :	Brief description highlighting the product's features and specifications.

* **user_id** :	Unique identifier of the customer who reviewed the product.

* **user_name** :	Name of the customer who submitted the review.

* **review_id** :	Unique identifier assigned to each customer review.

* **review_title** :	Short title or heading of the customer review.

* **review_content** :	Detailed text of the customer's review.

* **img_link** :	URL of the product image.

* **product_link** :	URL of the product page on Amazon.
