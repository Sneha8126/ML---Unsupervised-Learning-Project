## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Fill missing values in rating_count with mode
df['rating_count'].fillna(df['rating_count'].mode()[0], inplace=True)

# Remove symbols from price and discount columns
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False)

df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False)

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False)

# Remove commas from rating_count
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '', regex=False)

# Convert columns to numeric datatype
df['discounted_price'] = pd.to_numeric(df['discounted_price'])
df['actual_price'] = pd.to_numeric(df['actual_price'])
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'])
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Verify changes
df.info()
````

### What all manipulations have you done and insights you found?

#### **Manipulations Performed**

* Identified missing values and filled the missing values in the rating_count column using the mode.
* Removed currency symbols (₹), commas (,), and percentage symbols (%) from numerical columns.
* Converted discounted_price, actual_price, discount_percentage, rating, and rating_count into numeric data types.
* Verified the updated data types to ensure the dataset was ready for further analysis.

#### **Insights**

* The dataset contains 1,465 Amazon products with pricing, discounts, ratings, and review information.
* Only the rating_count column had 2 missing values, indicating that the dataset is nearly complete.
* Several numerical columns were stored as text due to currency symbols and commas, so preprocessing was necessary.
* After cleaning and data type conversion, the dataset became suitable for exploratory data analysis and unsupervised learning.

### 2. Handling Outliers

````python
# Check Outliers
numerical_cols = [
    'actual_price',
    'discounted_price',
    'discount_percentage',
    'rating',
    'rating_count'
]

for col in numerical_cols:
    plt.figure(figsize=(5,3))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
````

````python
print("Before Outliers treatment : ")
print()
for col in numerical_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col} : {len(outliers)} outliers")
````

````python
# Handling Outliers & Outlier treatments

for col in numerical_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
````

````python
print("After Outliers treatment : ")
print()
for col in numerical_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col} : {len(outliers)} outliers")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

* Outliers were identified using the **Interquartile Range (IQR) method** for all numerical features.

* Instead of removing the outliers, I applied the capping (Winsorization) technique using the clip() function.

* The lower and upper limits were calculated as Q1 − 1.5 × IQR and Q3 + 1.5 × IQR, and any values outside these limits were capped to the respective boundary values.

* This approach reduces the impact of extreme values while preserving all observations in the dataset. It helps maintain the original data size and prevents outliers from disproportionately influencing the clustering algorithm.

#### 1. Expand Contraction

````python
!pip install contractions
````

````python
# Expand Contraction

import contractions

df['review_content'] = df['review_content'].apply(
    lambda x: contractions.fix(str(x))
)

df[['review_content']].head()
````

#### 2. Lower Casing

````python
# Lower Casing

# Convert text to lowercase

df['review_content'] = df['review_content'].str.lower()

df[['review_content']].head()
````

#### 3. Removing Punctuations

````python
# Remove Punctuations
import string

df['review_content'] = df['review_content'].apply(
    lambda x: ''.join(char for char in str(x) if char not in string.punctuation)
)

# Display result
df[['review_content']].head()
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
import re

# Function to clean text
def clean_text(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', str(text))  # Remove URLs
    text = re.sub(r'\d+', '', text)                           # Remove numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)                   # Remove special characters
    return text

# Apply on review_content
df['review_content'] = df['review_content'].apply(clean_text)

# Display result
df[['review_content']].head()
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove Stopwords

import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df['review_content'] = df['review_content'].apply(
    lambda x: ' '.join(word for word in str(x).split() if word not in stop_words)
)

df[['review_content']].head()
````

````python
# Remove White spaces
df['review_content'] = df['review_content'].str.strip()

df['review_content'] = df['review_content'].str.replace(r'\s+', ' ', regex=True)

df[['review_content']].head()
````

#### 6. Rephrase Text

````python
# Rephrase Text

df['review_content'] = df['review_content'].apply(
    lambda x: ' '.join(str(x).split()).capitalize()
)

# Display result
df[['review_content']].head()
````

#### 7. Tokenization

````python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')  # If using newer NLTK versions
````

````python
# Tokenization
from nltk.tokenize import word_tokenize
# Tokenize the review_content column
df['tokens'] = df['review_content'].apply(lambda x: word_tokenize(str(x)))

# Display original text and tokens
df[['review_content', 'tokens']].head()
````

#### 8. Text Normalization

````python
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
````

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['lemmatized_text'] = df['tokens'].apply(
    lambda words: [lemmatizer.lemmatize(word) for word in words]
)

df[['tokens', 'lemmatized_text']].head()
````

##### Which text normalization technique have you used and why?

**I used Lemmatization as the text normalization technique. Lemmatization converts words into their base or dictionary form while preserving their actual meaning. For example, "products" becomes "product" and "batteries" becomes "battery". I chose Lemmatization because it produces meaningful root words, reduces vocabulary size, and improves the quality of text preprocessing for further NLP tasks such as POS tagging and TF-IDF vectorization.**
